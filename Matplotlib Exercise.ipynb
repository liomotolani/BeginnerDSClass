```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\USER\Downloads\company_sales_data.csv")
df
```

```python
#To check the columns in the dataset
for col_name in df:
    print(col_name)
```

```python
plt.plot(df['month_number'],df['total_profit'],linewidth=3)
plt.xticks(np.arange(min(df['month_number']), max(df['month_number'])+1, 1.0))
plt.title('Company Profit Per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.show()
```

```python
plt.plot(df['month_number'],df['facecream'],color='green',linewidth=3,marker='o', label = 'Facecream Sales Data')
plt.plot(df['month_number'],df['facewash'],color='red',linewidth=3,marker='o',label = 'Facewash Sales Data')
plt.plot(df['month_number'],df['toothpaste'],color='yellow',linewidth=3,marker='o', label = 'ToothPaste Sales Data')
plt.plot(df['month_number'],df['bathingsoap'],color='blue',linewidth=3,marker='o',label = 'Bathingsoap Sales Data')
plt.plot(df['month_number'],df['shampoo'],color='purple',linewidth=3,marker='o', label = 'Shampoo Sales Data')
plt.plot(df['month_number'],df['moisturizer'],color='brown',linewidth=3,marker='o',label = 'Moisturizer Sales Data')
plt.xticks(np.arange(min(df['month_number']), max(df['month_number'])+1, 1.0))
plt.title('Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Unit in Number')
plt.legend()
plt.show()
```

```python
plt.scatter(df['month_number'],df['toothpaste'],label= 'Tooth paste sales data')
plt.xticks(np.arange(min(df['month_number']), max(df['month_number'])+1, 1.0))
plt.title('Tooth Paste Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.grid(linestyle = '--')
plt.legend()
plt.show()
```

```python
w = 0.4
bar1 = np.arange(len(df['month_number']))
bar2 = [i + w for i in bar1]
plt.bar(bar1,df['facecream'],w,color='blue', label = 'Facecream Sales Data')
plt.bar(bar2,df['facewash'],w,color='red',label = 'Facewash Sales Data')
plt.xticks(np.arange(min(bar1), max(bar1)+1, 1.0))
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.grid(linestyle = '--')
plt.legend()
plt.show()
```

```python
sum_facecream = df['facecream'].sum()
sum_facewash = df['facewash'].sum()
sum_toothepaste = df['toothpaste'].sum()
sum_bathingsoap= df['bathingsoap'].sum()
sum_shampoo = df['shampoo'].sum()
sum_moisturizer = df['moisturizer'].sum()
totalsale_unit = [sum_facecream,sum_facewash,sum_toothepaste,sum_bathingsoap,sum_shampoo,sum_moisturizer]
pie_labels = 'FaceCream','FaceWash','ToothPaste','BathingSoap','Shampoo','Moisturizer'
plt.pie(totalsale_unit,labels = pie_labels,autopct='%1.1f%%')
plt.legend(loc = 'lower left',bbox_to_anchor=(0.8, -0.10))
plt.title('Sales Data')
plt.show()
```
