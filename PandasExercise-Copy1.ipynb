```python
# Question 9
germanCars = pd.DataFrame({'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]})
japaneseCars = pd.DataFrame({'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]})
concat_both_cars= pd.concat([germanCars,japaneseCars], keys = ["German", "Japan"])
print(concat_both_cars)
```

```python
# Question 10
car_price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
df_car_price = pd.DataFrame(car_price)
df_car_horsepower = pd.DataFrame(car_horsepower)

merged = pd.merge(df_car_price,df_car_horsepower,on ='Company')

print(merged)
```
